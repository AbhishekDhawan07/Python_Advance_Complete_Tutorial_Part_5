## Selection Sort Algorithm in Python

- Finds the minimum element repeatedly and places it at the correct position.
  

## Method - 1: Basic Selection Sort[ O(n²) ]

- Standard approach: find minimum in unsorted part, swap with first unsorted element.


```python
def selection_sort_basic(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i                          # assume current is minimum
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j                 # found new minimum
        arr[i], arr[min_idx] = arr[min_idx], arr[i]   # swap
    return arr
 
arr = [64, 25, 12, 22, 11]
print(selection_sort_basic(arr))   # [11, 12, 22, 25, 64]

```

## Method 2 — Selection Sort with Early Exit[ O(n²) but fewer swaps ]

- Skip the swap if the element is already in the correct position. Reduces unnecessary writes.


```python
def selection_sort_optimized(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:                     # only swap if needed
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
 
arr = [64, 25, 12, 22, 11]
print(selection_sort_optimized(arr))   # [11, 12, 22, 25, 64]
```

## Method 3 — Bidirectional Selection Sort (Cocktail Selection)[ O(n²) fewer passes]

- Find both minimum AND maximum in each pass. Places both at correct ends simultaneously — halves the number of passes.



```python
def selection_sort_bidirectional(arr):
    left = 0
    right = len(arr) - 1
    while left < right:
        min_idx = left
        max_idx = left
        for k in range(left, right + 1):     # find both min and max
            if arr[k] < arr[min_idx]:
                min_idx = k
            if arr[k] > arr[max_idx]:
                max_idx = k
        arr[left], arr[min_idx] = arr[min_idx], arr[left]
        if max_idx == left:                  # max was at left, now moved
            max_idx = min_idx
        arr[right], arr[max_idx] = arr[max_idx], arr[right]
        left  += 1
        right -= 1
    return arr
 
arr = [64, 25, 12, 22, 11]
print(selection_sort_bidirectional(arr))   # [11, 12, 22, 25, 64]

```

## Insertion Sort

- Builds the sorted array one element at a time by inserting each element into its correct position.


## Method 1 — Basic Insertion Sort[ O(n²) ]
- Compare each element with the previous ones and shift until the correct position is found.



```python
def insertion_sort_basic(arr):
    for i in range(1, len(arr)):
        key = arr[i]                         # element to be placed
        j = i - 1
        while j >= 0 and arr[j] > key:       # shift larger elements right
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key                     # insert at correct position
    return arr
 
arr = [12, 11, 13, 5, 6]
print(insertion_sort_basic(arr))   # [5, 6, 11, 12, 13]

```

## Method 2 — Insertion Sort with Binary Search [ O(n log n) comparisons, O(n²) shifts ]

- Use binary search to find the correct position instead of linear scan. Reduces comparisons from O(n) to O(log n) per element.


```python
import bisect
 
def insertion_sort_binary(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        # binary search finds correct position in O(log n)
        pos = bisect.bisect_left(arr, key, 0, i)
        arr[pos + 1 : i + 1] = arr[pos : i]  # shift elements right
        arr[pos] = key                        # insert key
    return arr
 
arr = [12, 11, 13, 5, 6]
print(insertion_sort_binary(arr))   # [5, 6, 11, 12, 13]

```

## Method 3 — Shell Sort (Improved Insertion Sort)[ O(n log² n)]

- Run insertion sort with a large gap first, then reduce the gap. Elements move to correct position faster because they travel in larger steps.


```python
def shell_sort(arr):
    n = len(arr)
    gap = n // 2
    while gap > 0:
        for i in range(gap, n):
            key = arr[i]
            j = i
            while j >= gap and arr[j - gap] > key:
                arr[j] = arr[j - gap]
                j -= gap
            arr[j] = key
        gap //= 2
    return arr              # ← this was missing

arr = [12, 11, 13, 5, 6]
print(shell_sort(arr))      # [5, 6, 11, 12, 13]
```

## Bubble Sort
- Repeatedly compares adjacent elements and swaps them if they are in the wrong order. Largest elements bubble up to the end.



## Method 1 — Basic Bubble Sort[ O(n²) ]

- Classic approach — always runs all passes regardless of whether the array is already sorted.


```python
def bubble_sort_basic(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):        # last i elements already sorted
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]   # swap
    return arr
 
arr = [64, 34, 25, 12, 22, 11, 90]
print(bubble_sort_basic(arr))   # [11, 12, 22, 25, 34, 64, 90]


```

## Method 2 — Optimized Bubble Sort with Early Exit[ O(n) best case ]
- Add a flag to detect if no swap happened in a pass. If no swap — array is already sorted — stop early. Best case becomes O(n).



```python
def bubble_sort_optimized(arr):
    n = len(arr)
    for i in range(n):
        swapped = False                      # track if any swap happened
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:                      # no swap = already sorted
            break
    return arr
 
arr = [64, 34, 25, 12, 22, 11, 90]
print(bubble_sort_optimized(arr))   # [11, 12, 22, 25, 34, 64, 90]


```

## Quick Sort
- Picks a pivot, partitions the array around it (smaller left, larger right), then recursively sorts both halves.


## Method 1 — Basic Quick Sort (Last Element as Pivot)   [ O(n²) worst case ]
- Always picks the last element as pivot. Fast in practice but degrades to O(n²) on already-sorted arrays.



```python
def quick_sort_basic(arr, s, e):
    if s >= e:
        return
    pivot = arr[e]                           # last element as pivot
    i = s - 1
    for j in range(s, e):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[e] = arr[e], arr[i + 1] # place pivot correctly
    pi = i + 1
    quick_sort_basic(arr, s, pi - 1)
    quick_sort_basic(arr, pi + 1, e)
 
arr = [10, 7, 8, 9, 1, 5]
quick_sort_basic(arr, 0, len(arr) - 1)
print(arr)   # [1, 5, 7, 8, 9, 10]

```

## Method 2 — Quick Sort with Random Pivot[ O(n log n) expected]
- Randomly select the pivot before each partition. This avoids the worst case on already-sorted or reverse-sorted arrays — making O(n²) extremely unlikely.


```python
import random
 
def quick_sort_random(arr, s, e):
    if s >= e:
        return
    # Randomly pick pivot and swap to end
    rand_idx = random.randint(s, e)
    arr[rand_idx], arr[e] = arr[e], arr[rand_idx]
    pivot = arr[e]
    i = s - 1
    for j in range(s, e):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[e] = arr[e], arr[i + 1]
    pi = i + 1
    quick_sort_random(arr, s, pi - 1)
    quick_sort_random(arr, pi + 1, e)
 
arr = [10, 7, 8, 9, 1, 5]
quick_sort_random(arr, 0, len(arr) - 1)
print(arr)   # [1, 5, 7, 8, 9, 10]
```

## Method 3 — Quick Sort with Median-of-Three Pivot[ O(n log n) better pivot ]

- Pick the median of first, middle, and last elements as pivot. Much better pivot choice — avoids worst case on sorted arrays without random number overhead.

```python
def median_of_three(arr, s, e):
    mid = (s + e) // 2
    # Sort s, mid, e positions and use middle as pivot
    if arr[s] > arr[mid]:
        arr[s], arr[mid] = arr[mid], arr[s]
    if arr[s] > arr[e]:
        arr[s], arr[e]   = arr[e],   arr[s]
    if arr[mid] > arr[e]:
        arr[mid], arr[e] = arr[e],   arr[mid]
    arr[mid], arr[e] = arr[e], arr[mid]      # move median to end
    return arr[e]
 
def quick_sort_median(arr, s, e):
    if s >= e:
        return
    pivot = median_of_three(arr, s, e)
    i = s - 1
    for j in range(s, e):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[e] = arr[e], arr[i + 1]
    pi = i + 1
    quick_sort_median(arr, s, pi - 1)
    quick_sort_median(arr, pi + 1, e)
 
arr = [10, 7, 8, 9, 1, 5]
quick_sort_median(arr, 0, len(arr) - 1)
print(arr)   # [1, 5, 7, 8, 9, 10]
```

## Merge Sort

- Divides the array into two halves, recursively sorts each half, then merges them back in sorted order.



## Method 1 — Basic Merge Sort[ O(n log n) always ]
- Classic top-down recursive approach. Always O(n log n) regardless of input — the most consistent sorting algorithm.


```python
def merge(arr, s, m, e):
    left  = arr[s : m + 1]                  # copy left half
    right = arr[m + 1 : e + 1]              # copy right half
    i = j = 0
    k = s
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            arr[k] = left[i];  i += 1
        else:
            arr[k] = right[j]; j += 1
        k += 1
    while i < len(left):                    # remaining left elements
        arr[k] = left[i];  i += 1;  k += 1
    while j < len(right):                   # remaining right elements
        arr[k] = right[j]; j += 1;  k += 1
 
def merge_sort_basic(arr, s, e):
    if s >= e:
        return
    m = s + (e - s) // 2
    merge_sort_basic(arr, s, m)
    merge_sort_basic(arr, m + 1, e)
    merge(arr, s, m, e)
 
arr = [38, 27, 43, 3, 9, 82, 10]
merge_sort_basic(arr, 0, len(arr) - 1)
print(arr)   # [3, 9, 10, 27, 38, 43, 82]

```

## Method 2 — Merge Sort with Early Skip[ O(n log n) faster on partial sorted ]
- Before merging, check if arr[mid] <= arr[mid+1]. If yes — both halves are already in order — skip the merge entirely. Huge speedup on nearly-sorted data.



```python
def merge_sort_skip(arr, s, e):
    if s >= e:
        return
    m = s + (e - s) // 2
    merge_sort_skip(arr, s, m)
    merge_sort_skip(arr, m + 1, e)
    # Skip merge if already in order
    if arr[m] <= arr[m + 1]:
        return                               # no merge needed
    merge(arr, s, m, e)                      # same merge function as above
 
arr = [38, 27, 43, 3, 9, 82, 10]
merge_sort_skip(arr, 0, len(arr) - 1)
print(arr)   # [3, 9, 10, 27, 38, 43, 82]

```

## Method 3 — Bottom-Up Iterative Merge Sort[ O(n log n), no recursion stack ]
- Instead of recursing top-down, start merging pairs of size 1, then size 2, then size 4 and so on. No recursion stack — saves O(log n) stack space. Better for very large arrays.

```python
def merge_sort_iterative(arr):
    n = len(arr)
    size = 1                                 # start with subarrays of size 1
    while size < n:
        for s in range(0, n, size * 2):      # pick each pair of subarrays
            m = min(s + size - 1, n - 1)
            e = min(s + size * 2 - 1, n - 1)
            if m < e:                        # only merge if right half exists
                merge(arr, s, m, e)          # same merge function as above
        size *= 2                            # double the subarray size
    return arr
 
arr = [38, 27, 43, 3, 9, 82, 10]
print(merge_sort_iterative(arr))   # [3, 9, 10, 27, 38, 43, 82]

```
